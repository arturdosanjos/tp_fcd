```python
# Instalação de pacotes auxiliares (descomente se rodar no Colab)
!pip install shap scikit-learn

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import shap
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split

# Configuração visual do Seaborn
sns.set_theme(style="whitegrid")

# 1. Carregamento dos dados
url_torneios = "https://raw.githubusercontent.com/jfjelstul/worldcup/master/data-csv/tournaments.csv"
url_partidas = "https://raw.githubusercontent.com/jfjelstul/worldcup/master/data-csv/matches.csv"

df_torneios = pd.read_csv(url_torneios)
df_partidas = pd.read_csv(url_partidas)

# 2. Estruturação e Limpeza (Merge)
df_partidas['match_date'] = pd.to_datetime(df_partidas['match_date'])
df_partidas['total_goals'] = df_partidas['home_team_score'] + df_partidas['away_team_score']
df_partidas['goal_difference'] = abs(df_partidas['home_team_score'] - df_partidas['away_team_score'])

# Trazendo o país sede e o ano do torneio para a base de partidas
df_consolidado = pd.merge(
    df_partidas,
    df_torneios[['tournament_id', 'year', 'host_country', 'winner']],
    on='tournament_id',
    how='left'
)

print("Setup concluído! Base pronta para análise.")

display(df_consolidado.head())
```

Muitas análises de futebol exigem olhar para o desempenho individual das equipes (independente de serem mandantes ou visitantes). Vamos criar um DataFrame auxiliar df_equipes para facilitar isso:



```python
# --- CÉLULA DE PREPARAÇÃO PARA A PARTE 2 ---
import pandas as pd
import numpy as np

# Desempilhando mandantes e visitantes para ter uma linha por equipe por jogo
df_mandantes = df_consolidado[['match_id', 'year', 'stage_name', 'home_team_name', 'home_team_score', 'away_team_score', 'host_country']].copy()
df_mandantes.columns = ['match_id', 'year', 'stage_name', 'equipe', 'gols_feitos', 'gols_sofridos', 'host_country']
df_mandantes['vitoria'] = df_mandantes['gols_feitos'] > df_mandantes['gols_sofridos']
df_mandantes['empate'] = df_mandantes['gols_feitos'] == df_mandantes['gols_sofridos']
df_mandantes['mando'] = 'Casa'

df_visitantes = df_consolidado[['match_id', 'year', 'stage_name', 'away_team_name', 'away_team_score', 'home_team_score', 'host_country']].copy()
df_visitantes.columns = ['match_id', 'year', 'stage_name', 'equipe', 'gols_feitos', 'gols_sofridos', 'host_country']
df_visitantes['vitoria'] = df_visitantes['gols_feitos'] > df_visitantes['gols_sofridos']
df_visitantes['empate'] = df_visitantes['gols_feitos'] == df_visitantes['gols_sofridos']
df_visitantes['mando'] = 'Fora'

df_equipes = pd.concat([df_mandantes, df_visitantes], ignore_index=True)
print("DataFrame 'df_equipes' criado com sucesso para apoiar as análises!")
```

```python
#questao 01

def padronizar_jogo(nome_jogo:str):
  nome_jogo = nome_jogo.split(" vs ")
  nome_jogo_ordenado = sorted(nome_jogo)

  return " vs ".join(nome_jogo_ordenado)

df_consolidado_padronizado = df_consolidado.copy()
df_consolidado_padronizado['match_name'] = df_consolidado_padronizado['match_name'].apply(padronizar_jogo)

# Pega o nome da rivalidade mais frequente
rivalidade_his = df_consolidado_padronizado['match_name'].value_counts().index[0]

# 1. Pega apenas a lista (Série) de 'match_id' que pertencem a essa rivalidade
ids_da_rivalidade = df_consolidado_padronizado[df_consolidado_padronizado["match_name"] == rivalidade_his]["match_id"]

# 2. Filtra o df_equipes trazendo apenas as linhas cujo 'match_id' ESTEJA NA lista acima
resultado = df_equipes[df_equipes['match_id'].isin(ids_da_rivalidade)]

vitorias_por_equipe = resultado[resultado['vitoria'] == True]['equipe'].value_counts()
vitorias_por_equipe
```

```python
# questão 13
# Violin Plot (Gráfico de Violino): 
# Crie um gráfico de violino comparando a distribuição de gols marcados por partida pelas 4 seleções mais vencedoras 
# (Brasil, Alemanha, Itália, Argentina). Quem é mais regular e quem oscila mais?

home_goals = df_consolidado[['home_team_name', 'home_team_score']].copy()
home_goals.columns = ['Selecao', 'Gols']

away_goals = df_consolidado[['away_team_name', 'away_team_score']].copy()
away_goals.columns = ['Selecao', 'Gols']

# 2. Unir tudo em um único DataFrame e garantir que Gols seja numérico
df_gols = pd.concat([home_goals, away_goals], ignore_index=True)
df_gols['Gols'] = pd.to_numeric(df_gols['Gols'], errors='coerce')

selecoes_alvo = ['Brazil', 
                 'Germany', 'East Germany', 'West Germany', 'Germany FR', # alemanha teve vários nomes
                 'Italy',
                 'Argentina']

df_top4 = df_gols[df_gols['Selecao'].isin(selecoes_alvo)].copy()

# Padronizar os nomes para o gráfico ficar limpo (caso haja variações)
mapeamento_nomes = {
    'Brazil': 'Brasil',
    'Germany': 'Alemanha', 'West Germany': 'Alemanha', 'Germany FR': 'Alemanha', 'East Germany': 'Alemanha',
    'Italy': 'Itália',
    'Argentina': 'Argentina'
}
df_top4['Selecao'] = df_top4['Selecao'].replace(mapeamento_nomes)

# 4. Criar o Gráfico de Violino
plt.figure(figsize=(10, 6))
sns.violinplot(data=df_top4, x='Selecao', y='Gols', hue='Selecao', palette='Set2', inner='quartile', legend=False)

# 5. Ajustes visuais
plt.title('Distribuição de Gols por Partida: Brasil, Alemanha, Itália e Argentina')
plt.xlabel('Seleção')
plt.ylabel('Gols Marcados na Partida')

# Exibir o gráfico
plt.show()

# Calcular métricas de dispersão para as 4 seleções
metricas = df_top4.groupby('Selecao')['Gols'].agg(
    Média='mean',
    Desvio_Padrão='std', # Medida principal de oscilação
    Mínimo='min',
    Máximo='max' # Ajuda a ver a amplitude
).round(2) # Arredondar para 2 casas decimais

# Calcular a Amplitude (Máximo - Mínimo)
metricas['Amplitude'] = metricas['Máximo'] - metricas['Mínimo']

# Ordenar pelo Desvio Padrão (do menor para o maior)
metricas = metricas.sort_values(by='Desvio_Padrão')

print("Análise de Regularidade e Oscilação:")
print(metricas)
```

```python
# questão 15
#  Crie uma tabela pivô onde as linhas são as fases do torneio (Grupos, Oitavas, Quartas, Semi, Final)
# e as colunas são as edições (Anos). Preencha com a média de gols. O heatmap mostrará se os jogos ficam 
# mais "truncados" (frios) à medida que chegam na final.

df_analise = df_consolidado.copy()

# padronizando
mapeamento_fases = {
    'group stage': 'Fase de grupos', 'second group stage':'Fase de grupos',
    'round of 16': 'Oitavas',
    'quarter-finals': 'Quartas','quarter-final': 'Quartas',
    'third-place match':'Partida do terceiro lugar',
    'semi-finals': 'Semi-final','semi-final':'Semi-final',
    'final': 'Final', 'final round':'Final'
}

df_analise['fase_mapeada'] = df_analise['stage_name'].replace(mapeamento_fases)

# definindo ordem cronológica das fases
fases_ordenadas = ['Fase de grupos', 'Oitavas', 'Quartas', 'Semi-final','Partida do terceiro lugar', 'Final']

# filtrado df
df_filtrado = df_analise[df_analise['fase_mapeada'].isin(fases_ordenadas)].copy()

# tabela de pivô com média dos gols
tabela_pivo = df_filtrado.pivot_table(
    index='fase_mapeada', 
    columns='year', 
    values='total_goals', 
    aggfunc='mean'
)

# reordenar as linhas conforme a ordem natural do campeonato
tabela_pivo = tabela_pivo.reindex(fases_ordenadas)

# gerando gráfico de calor
plt.figure(figsize=(16, 7))

sns.heatmap(tabela_pivo, 
            annot=True,      
            fmt=".2f",       
            cmap="YlOrRd",   
            linewidths=.5,   
            cbar_kws={'label': 'Média de Gols por Partida'})

plt.title('Heatmap: Média de Gols por Fase e Ano de Copa do Mundo', fontsize=14, pad=15)
plt.xlabel('Ano da Edição', fontsize=12)
plt.ylabel('Fase do Torneio', fontsize=12)

plt.tight_layout()
plt.show()
```

```python
# questao 17
# Vamos calcular a correlação entre o passar do tempo (Ano da Copa) e o Equilíbrio (Diferença de Gols)
from scipy.stats import spearmanr

df_spearman = df_consolidado.copy()


# 3. Calcular a Correlação de Spearman e o P-valor
coeficiente, p_valor = spearmanr(df_spearman['year'], df_spearman['goal_difference'])

print("=== Análise de Correlação de Spearman ===")
print(f"Coeficiente de Correlação: {coeficiente:.4f}")
print(f"P-valor: {p_valor:.4e}")

# Interpretação estatística automática
if p_valor < 0.05:
    print("O resultado é estatisticamente significante (p < 0.05).")
    if coeficiente < 0:
        print("Tendência: O futebol está ficando mais equilibrado com o tempo.")
    else:
        print("Tendência: A disparidade de gols aumentou com o tempo.")
else:
    print("O resultado NÃO é estatisticamente significante (p >= 0.05). Não há tendência clara de mudança no equilíbrio.")

# 4. Gráfico de Dispersão com Linha de Tendência Logística/Linear
plt.figure(figsize=(12, 6))

# Agrupar pela média da diferença de gols por ano para o gráfico ficar mais limpo visualmente
df_agrupado_ano = df_spearman.groupby('year')['goal_difference'].mean().reset_index()

sns.regplot(data=df_agrupado_ano, x='year', y='goal_difference', 
            color='darkslateblue', scatter_kws={'s': 50}, line_kws={'color': 'crimson'})

plt.title('Evolução da Média da Diferença de Gols por Partida ao Longo dos Anos')
plt.xlabel('Ano da Copa')
plt.ylabel('Média da Diferença de Gols')
plt.grid(True, linestyle='--', alpha=0.6)

plt.show()
```

```python
# questao 19
# Gols no futebol costumam modelar perfeitamente uma Distribuição de Poisson. C
# Calcule o Valor Esperado ($\lambda$) de gols por partida em toda a história. Usando esse $\lambda$, 
# qual é a probabilidade teórica de um jogo terminar em 0x0? Como essa probabilidade matemática se ]
# compara com a porcentagem real de jogos que terminaram 0x0 na sua base?

import math

gols_validos = df_consolidado['total_goals'].dropna()

# calculando o valor esperado (Lambda): A média de gols por partida
lambda_gols = gols_validos.mean()

# calculando probabilidade através da fórmula, como k=0 ela fica simplificada
prob_teorica_0x0 = math.exp(-lambda_gols)

# calculando a porcentagem real de jogos 0x0 na base
total_jogos = len(gols_validos)
jogos_0x0 = len(gols_validos[gols_validos == 0])
prob_real_0x0 = jogos_0x0 / total_jogos

# exibindo os resultados
print("=== Análise de Poisson: A Realidade do 0x0 ===")
print(f"Valor Esperado (Lambda - Média de Gols): {lambda_gols:.4f}")
print(f"Probabilidade Teórica (Matemática) de 0x0: {prob_teorica_0x0:.2%}")
print(f"Porcentagem Real de 0x0 na História: {prob_real_0x0:.2%}")

# comparação
diferenca = prob_real_0x0 - prob_teorica_0x0
if diferenca > 0:
    print(f"\nConclusão: Ocorrem {diferenca:.2%} MAIS jogos 0x0 na realidade do que a matemática prevê.")
else:
    print(f"\nConclusão: Ocorrem {abs(diferenca):.2%} MENOS jogos 0x0 na realidade do que a matemática prevê.")
```

```python
# questão 21
# Usaremos o Teste T (Welch) para verificar se há uma diferença estatisticamente significativa na média de gols quando o país anfitrião está em campo.

# Hipótese Nula ($H_0$): A média de gols é igual nos dois cenários. O anfitrião não muda o padrão do jogo.
# Hipótese Alternativa ($H_1$): A média de gols é diferente. O anfitrião impacta o número de gols.

from scipy.stats import ttest_ind

# Grupo A: Jogos onde o anfitrião está em campo (como mandante ou visitante)
filtro_anfitriao = (df_consolidado['home_team_name'] == df_consolidado['host_country']) | \
                   (df_consolidado['away_team_name'] == df_consolidado['host_country'])

jogos_com_anfitriao = df_consolidado[filtro_anfitriao]

# Grupo B: Jogos neutros (sem o anfitrião)
jogos_sem_anfitriao = df_consolidado[~filtro_anfitriao]

# 2. Extrair apenas a coluna de gols (já sabemos que é numérica graças ao seu ajuste anterior)
gols_anfitriao = jogos_com_anfitriao['total_goals'].dropna()
gols_normais = jogos_sem_anfitriao['total_goals'].dropna()

# 3. Aplicar o Teste T de Welch (o parâmetro equal_var=False é o que define o método de Welch)
t_stat, p_valor = ttest_ind(gols_anfitriao, gols_normais, equal_var=False)

# 4. Calcular médias para visualização
media_anfitriao = gols_anfitriao.mean()
media_normais = gols_normais.mean()

# 5. Exibir os Resultados
print("=== Teste de Hipótese: O Efeito do País Anfitrião ===")
print(f"Amostra 1 (Com Anfitrião): {len(gols_anfitriao)} jogos | Média de Gols: {media_anfitriao:.2f}")
print(f"Amostra 2 (Sem Anfitrião): {len(gols_normais)} jogos | Média de Gols: {media_normais:.2f}")
print("-" * 45)
print(f"Estatística T: {t_stat:.4f}")
print(f"P-valor: {p_valor:.4e}")
print("-" * 45)

# Interpretação
alfa = 0.05
if p_valor < alfa:
    print("Conclusão: Rejeitamos a Hipótese Nula (p < 0.05).")
    print("Resultado: Existe uma diferença estatisticamente significativa na média de gols quando o país sede está jogando.")
else:
    print("Conclusão: Falhamos em rejeitar a Hipótese Nula (p >= 0.05).")
    print("Resultado: NÃO há evidências estatísticas suficientes para afirmar que a presença do anfitrião altera o padrão médio de gols do torneio.")
```
